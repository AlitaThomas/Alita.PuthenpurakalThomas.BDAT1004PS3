### Question 7

Create a vertical bar chart comparing the number of marriages and divorces per
capita in the U.S. between 1900, 1950, and 2000.
Don't forget to label your axes!

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv('us-marriages-divorces-1867-2014.csv')
data = df[(df.Year == 2000) | (df.Year == 1900) | (df.Year == 1950)]
data.plot(kind= 'bar', x='Year', y=['Marriages_per_1000','Divorces_per_1000'], label=['Marriages per capita','Divorces per capita'])
plt.title("Create a vertical bar \n chart comparing the number \n of marriages and divorces \n per capita in the \n U.S. between 1900, 1950, and 2000.")
plt.legend(loc='best')
plt.show()
```
