```python
import pandas as pd
import numpy as np
url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/02_Filtering_%26_Sorting/Euro12/Euro_2012_stats_TEAM.csv'
euro12 = pd.read_csv(url, error_bad_lines=False)
```

```python
print('------- Print Table ------')
print(euro12)
```

```python

print('------- Select only the Goal column ------')
euro12['Goals']
print(euro12['Goals'])
```

```python

print('------- How many team participated in the Euro2012 ------')
euro12['Team'].agg('count')
print(euro12['Team'].agg('count'))

```

```python
print('------- What is the number of columns in the dataset ------')
len(euro12.columns)
print(len(euro12.columns))

```

```python

print('------- View only the columns Team, Yellow Cards and Red Cards and assign them to a dataframe called discipline ------')
discipline = euro12[['Team', 'Yellow Cards', 'Red Cards']]
print(discipline)
```

```python
print('------- Sort the teams by Red Cards ------')
euro12 = euro12.sort_values(by ='Red Cards')
print(euro12)
print('------- Sort the teams by Yellow Cards ------')
euro12 = euro12.sort_values(by ='Yellow Cards')
print(euro12)
```

```python

print('------- Calculate the mean Yellow Cards given per Team ------')
euro12['Yellow Cards'].mean()
print(euro12['Yellow Cards'].mean())
```

```python

print('------- teams that scored more than 6 goal ------')
euro12[euro12.Goals > 6]
print(euro12[euro12.Goals > 6])
```

```python
print('------- Select the teams that start with G ------')
euro12[euro12.Team.str.startswith("G")]
print(euro12[euro12.Team.str.startswith("G")])
```

```python
print('------- Select the first 7 columns ------')
euro12.iloc[: , 0:7]
print(euro12.iloc[: , 0:7])
```

```python
print('------- Select all columns except the last 3 ------')
euro12.iloc[: , :-3]
print(euro12.iloc[: , :-3])
```

```python
euro12.loc[euro12.Team.isin(['England', 'Italy', 'Russia']), ['Team','Shooting Accuracy']]
```
