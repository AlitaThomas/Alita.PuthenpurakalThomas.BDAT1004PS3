### Question 6

Create a line plot showing the number of marriages and divorces per capita in the
U.S. between 1867 and 2014. Label both lines and show the legend.
Don't forget to label your axes!


```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv('us-marriages-divorces-1867-2014.csv')
df.plot(x='Year', y=['Marriages_per_1000','Divorces_per_1000'], label=['Marriages per capita','Divorces per capita'])
plt.xlim(1867, 2014)
plt.title('The number of \n marriages and divorces per capita \n in the U.S. between 1867 and 2014')
plt.legend(loc='best')
plt.show()
```
